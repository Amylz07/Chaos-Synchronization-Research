```python
# ordpy code for ordinal distribution
sigma = 16 #original: 16
b = 4 #original: 4
tmax = 100
r = 80

t = np.arange(0.0, tmax, 0.1)

def two_lorenz_odes(X, t):
    x1, x2, x3, y2, y3 = X
    dx1 = sigma*(x2-x1)
    dx2 = -x1*x3+r*x1-x2
    dx3 = x1*x2-b*x3
    dy2 = -x1*y3+r*x1-y2
    dy3 = x1*y2-b*y3
    return (dx1, dx2, dx3, dy2, dy3)

y0 = [15, 20, 30, 10, 20]
X = odeint(two_lorenz_odes, y0, t, rtol=1e-6)
x1, x2, x3, y2, y3 = X.T # unpack columns

OD_x1 = ordpy.ordinal_distribution(x1, dx=4, dy=1, taux=1, tauy=1, return_missing=False, tie_precision=None)
OP_x1 = OD_x1[1][0]
OD_x2 = ordpy.ordinal_distribution(x2, dx=4, dy=1, taux=1, tauy=1, return_missing=False, tie_precision=None)
OP_x2 = OD_x2[1][0]
OD_x3 = ordpy.ordinal_distribution(x3, dx=4, dy=1, taux=1, tauy=1, return_missing=False, tie_precision=None)
OP_x3 = OD_x3[1][0]
OD_y2 = ordpy.ordinal_distribution(y2, dx=4, dy=1, taux=1, tauy=1, return_missing=False, tie_precision=None)
OP_y2 = OD_y2[1][0]
OD_y3 = ordpy.ordinal_distribution(y3, dx=4, dy=1, taux=1, tauy=1, return_missing=False, tie_precision=None)
OP_y3 = OD_y3[1][0]


ordinal_patterns=[]
probabilities=[]
i=0
while i<len(OD_x1[0]):
    ordinal_patterns.append(OD_x1[0][i])
    probabilities.append(OD_x1[1][i])
    i+=1
# print(OD_x1[1][0]-OD_x1[1][len(OD_x1[0])-1])

data={"ordinal patterns":ordinal_patterns, "probabilities":probabilities}

dataframe = pd.DataFrame(data)
print("dataframe:{}".format(dataframe))

def my_autopct(pct):
    return ('%.2f' % pct) if pct > 1 else ''

fig = plt.figure(figsize =(10, 7))
plt.pie(probabilities, autopct=my_autopct, labels = ordinal_patterns)
plt.legend(ordinal_patterns, loc="center right")
plt.title("r={}".format(r))

figure(figsize=(20,10))
plt.plot(t, x1)
plt.xlabel('t')
plt.ylabel('x1')
plt.title("r={}".format(r))
```
